```python
import pandas as pds
data = pds.read_csv("AAPL.csv")
print(data.tail())

data = data[-252:]
print(len(data))
```

```python
#Garman Klass volatility calculation
import math
#N is # of trading days in a year
#n is the chosen sample size
def Garman_Klass(data):
    n = len(data)
    N = 252
    scaling_factor = N/n
    totalSum = 0
    currentSum = 0
    for x in range(len(data)):
        currentSum = (1/2) * ((math.log10(data.iloc[x]['High']/data.iloc[x]['Low'])) ** 2) - (2 * math.log10(2) - 1) * (math.log10((data.iloc[x]['Close'])/(data.iloc[x]['Open'])) ** 2)
        totalSum += currentSum
    returnVal = scaling_factor * totalSum
    returnVal = returnVal ** 0.5
    return returnVal

print(Garman_Klass(data))
```
